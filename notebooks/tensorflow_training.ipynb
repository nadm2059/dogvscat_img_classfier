```python
# Import the sys module to manipulate the Python runtime environment
import sys
import numpy as np

# Import the os module for interacting with the operating system
import os

# Add the parent directory to the system path so Python can find modules in that directory
sys.path.append(os.path.abspath(".."))

# Import functions to build and train the TensorFlow model from a custom module
from models.tensorflow_model import build_tf_model, train_tf_model

# Import a function to load and preprocess TensorFlow datasets
from utils.data_preprocessing import get_tf_datasets

# Import a function to visualize the training history (accuracy/loss over epochs)
from utils.visualization import plot_history, plot_confusion_matrix

# Define the path to the dataset directory
data_dir = r"C:\Multi-Framework Image Classification\data"

# Load the training and validation datasets using the custom preprocessing function
train_ds, val_ds = get_tf_datasets(data_dir)

# Build the TensorFlow model using a custom architecture defined in the module
model = build_tf_model()

# Train the model on the training dataset and validate it on the validation dataset
history = train_tf_model(model, train_ds, val_ds)

# Plot the training history (typically shows accuracy and/or loss for train and val sets)
plot_history(history)

# Confusion Matrix
y_true = []
y_pred = []

for images, labels in val_ds:
    predictions = model.predict(images)
    predicted_labels = np.argmax(predictions, axis=1)
    y_true.extend(labels.numpy())
    y_pred.extend(predicted_labels)

class_names = ['Cat', 'Dog']
plot_confusion_matrix(y_true, y_pred, class_names)
```
