```python
# Import necessary libraries
import os                    # For interacting with the file system (e.g., listing directory contents)
import matplotlib.pyplot as plt  # For plotting images using matplotlib
import random                # For selecting random samples from a list
from PIL import Image        # For opening and manipulating image files

# Define a function to display 5 random images from a given folder with a given label
def show_random_images(folder, label):
    files = os.listdir(folder)               # List all file names in the specified folder
    images = random.sample(files, 5)         # Randomly select 5 image filenames from the list
    fig, axs = plt.subplots(1, 5, figsize=(15, 3))  # Create a 1x5 grid of subplots for displaying the images
    for i, img_name in enumerate(images):    # Iterate through each selected image filename
        img = Image.open(os.path.join(folder, img_name))  # Open the image using its full path
        axs[i].imshow(img)                   # Display the image in the subplot
        axs[i].axis('off')                   # Turn off axis ticks and labels
        axs[i].set_title(label)              # Set the title of the subplot to the given label (e.g., "Cat" or "Dog")

# Show 5 random cat images from the training set with label "Cat"
show_random_images("C:/Multi-Framework Image Classification/data/train/cats", "Cat")

# Show 5 random dog images from the training set with label "Dog"
show_random_images("C:/Multi-Framework Image Classification/data/train/dogs", "Dog")


```
